# Data Analysis Workshop
Let's visualize a chess dataset!

```python
gamedata = pd.read_csv('./games.csv')
gamedata.shape # we have 20,000+ game data
```

```python
# Get number of unique opening names
len(gamedata['opening_name'].unique())
```

```python
openings = gamedata['opening_name']
```

```python
# Since there are so many variations for openings, 
# we try to generalize each opening name in the dataset to its base opening name
def openingTransform(val):
    if ':' in val:
        val = val[:val.index(':')]
    if '|' in val:
        val = val[:val.index('|')].strip()
    if 'Accepted' in val:
        val = val[:val.index('Accepted')].strip()
    else:
        if 'Declined' in val:
            val = val[:val.index('Declined')].strip()
    if '#' in val:
        val = val[:val.index('#')].strip()
    if 'Accelerated' in val:
        val = val[:val.index('Accelerated')].strip()
    return val
```

```python
gamedata['opening_transformed'] = [openingTransform(val) for val in gamedata['opening_name']]
```

```python
openingsUnique = gamedata['opening_transformed'].unique()
```

```python
# Here are the top openings used and the corresponding chess locations
topFive = gamedata['opening_transformed'].value_counts()[:6]
topFive
```
